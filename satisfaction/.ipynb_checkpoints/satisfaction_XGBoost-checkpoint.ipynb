```python
import import_ipynb
from satisfaction_data_preprocessing import get_satisfaction_dataset

X_train, X_test, y_train, y_test = get_satisfaction_dataset()
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import roc_auc_score
import warnings
warnings.filterwarnings('ignore')

xgb_clf = XGBClassifier(n_estimators=500, random_state=156)
xgb_clf.fit(X_train, y_train, early_stopping_rounds=100,
           eval_metric='auc', eval_set=[(X_train,y_train),(X_test,y_test)])

xgb_roc_score = roc_auc_score(y_test, xgb_clf.predict_proba(X_test)[:,1], average='macro')
print('ROC AUC :{0:.4f}'.format(xgb_roc_score))
```

#### AUC 가 0.84 가 나온것을 볼수 있다. 최적의 파라미터를 찾기위해 GridSearchCV 를 사용하고 빠른 탐색을 위해 n_estimators 는 100으로 설정한다.

```python
from sklearn.model_selection import GridSearchCV

xgb_clf = XGBClassifier(n_estimators=100)

params = {'max_depth':[5,7],
         'min_child_weight' : [1,3],
          'colsample_bytree':[0.5,0.75]
         }

gridcv = GridSearchCV(xgb_clf, param_grid=params, cv=3)
gridcv.fit(X_train,y_train, early_stopping_rounds=30, eval_metric="auc",
          eval_set=[(X_train,y_train),(X_test,y_test)])

print("GridSearchCV 최적 파라미터 :",gridcv.best_params_)

xgb_roc_score=roc_auc_score(y_test,gridcv.predict_proba(X_test)[:,1], average='macro')
print('ROC AUC:{0:.4f}'.format(xgb_roc_score))
```

#### 최적의 파라미터는 'colsample_bytree': 0.5, 'max_depth': 7, 'min_child_weight': 3 임을 알 수 있다.
n_estimators 는 1000으로 증가시키고, learning_rate 는 0.02로 감소시키고 ,reg_alpha=0.03 으로 설정

```python
xgb_clf = XGBClassifier(n_estimators=1000, random_state=156, learning_rate=0.02, colsample_bytree=0.5,
                       max_depth=7, min_child_weight=3, reg_alpha=0.03)

xgb_clf.fit(X_train,y_train, early_stopping_rounds=200, eval_metric="auc", eval_set=[(X_train,y_train),(X_test,y_test)])

xgb_roc_score = roc_auc_score(y_test, xgb_clf.predict_proba(X_test)[:,1], average='macro')
print("ROC AUC : {0:.4f}".format(xgb_roc_score))
```

#### ROC AUC : 0.8460 으로 전보다 조금 상승한 것을 알 수 있다. 피처의 중요도를 시각화 한다.

```python
from xgboost import plot_importance
import matplotlib.pyplot as plt
%matplotlib inline

fig, ax = plt.subplots(1,1,figsize=(10,8))
plot_importance(xgb_clf, ax=ax, max_num_features=20, height=0.4)
```
